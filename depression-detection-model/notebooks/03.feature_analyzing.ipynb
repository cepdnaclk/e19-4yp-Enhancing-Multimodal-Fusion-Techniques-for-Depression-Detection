```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pointbiserialr, zscore

print("=== Starting EDA script ===")

# ----------- STEP 1: LOAD AND INSPECT THE DATA -----------
try:
    df = pd.read_parquet("../data/processed/all_user_combined_data_processed.parquet")
    print(f"Data loaded successfully! Shape: {df.shape}")
except Exception as e:
    print(f"Failed to load data: {e}")
    exit(1)

print("\n--- Data Types ---")
print(df.dtypes)

print("\n--- First 5 rows ---")
print(df.head())

print("\n--- Statistical Summary ---")
print(df.describe(include='all'))

target_col = 'Depression_label'  
if target_col in df.columns:
    print(f"\n--- Target Variable ('{target_col}') Distribution ---")
    print(df[target_col].value_counts(normalize=True))
else:
    print(f"Target column '{target_col}' not found!")

# ----------- STEP 2: CHECK FOR MISSING DATA -----------
print("\n=== Missing Data ===")
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
print(missing)

# Visualize missing data heatmap with matplotlib only
plt.figure(figsize=(10, 6))
plt.imshow(df.isnull(), aspect='auto', cmap='viridis', interpolation='none')
plt.colorbar(label='Missing Value')
plt.title('Missing Values Heatmap')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.xticks(ticks=np.arange(len(df.columns)), labels=df.columns, rotation=90, fontsize=8)
plt.yticks([])  # hide row labels for clarity
plt.tight_layout()
plt.show()
print("Missing values heatmap shown.")

# ----------- STEP 3: STATISTICAL SUMMARY & TARGET BALANCE -----------
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
print("\n=== Numeric Features Summary ===")
print(df[numeric_cols].describe())

cat_cols = df.select_dtypes(include='object').columns.tolist()
print("\n=== Categorical Features Summary ===")
for col in cat_cols:
    print(f"\nValue counts for {col}:")
    print(df[col].value_counts())

# Plot target variable balance with matplotlib
if target_col in df.columns:
    counts = df[target_col].value_counts()
    plt.figure(figsize=(6,4))
    plt.bar(counts.index.astype(str), counts.values, color='skyblue')
    plt.title('Target Variable Distribution')
    plt.xlabel(target_col)
    plt.ylabel('Count')
    plt.show()
    print("Target variable distribution plot shown.")

# ----------- STEP 4: FEATURE RELATIONSHIPS (Target vs Features) -----------
print("\n=== Computing point-biserial correlations with target ===")

def compute_pointbiserial_correlations(df, feature_cols, label_col):
    correlations = {}
    for col in feature_cols:
        if col == label_col:
            continue
        try:
            if df[col].nunique() > 1:
                corr, pval = pointbiserialr(df[label_col], df[col])
                correlations[col] = (corr, pval)
        except Exception as e:
            print(f"Skipping {col} due to error: {e}")
    return pd.DataFrame.from_dict(correlations, orient='index', columns=['Correlation', 'p-value']).sort_values('Correlation', ascending=False)

if target_col in df.columns:
    feature_cols = [col for col in numeric_cols if col != target_col]
    pb_corr_df = compute_pointbiserial_correlations(df, feature_cols, target_col)
    print("\n=== Point-Biserial Correlations with Target ===")
    print(pb_corr_df.head(10))

# ----------- STEP 5: VISUALIZATIONS -----------
print("\n=== Plotting feature distributions by target (only 10 features) ===")
plot_features = feature_cols[:10]  # only first 10 numeric features

for col in plot_features:
    plt.figure(figsize=(8,4))
    for val in df[target_col].unique():
        subset = df[df[target_col] == val]
        plt.hist(subset[col].dropna(), bins=30, alpha=0.5, label=f'{target_col}={val}', density=True)
    plt.title(f'Distribution of {col} by {target_col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.legend()
    plt.tight_layout()
    plt.show()
print("Histogram plots done.")

print("\n=== Plotting boxplots by target (only 10 features) ===")
for col in plot_features:
    data_to_plot = [df[df[target_col] == val][col].dropna() for val in df[target_col].unique()]
    plt.figure(figsize=(6,4))
    plt.boxplot(data_to_plot, labels=[str(val) for val in df[target_col].unique()])
    plt.title(f'{col} by {target_col}')
    plt.xlabel(target_col)
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
print("Boxplots done.")

print("\n--- End of EDA ---")
```
