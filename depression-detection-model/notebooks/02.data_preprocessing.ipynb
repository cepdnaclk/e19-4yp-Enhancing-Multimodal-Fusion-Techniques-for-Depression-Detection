```python
import pandas as pd

df = pd.read_parquet("../data/processed/all_user_combined_data.parquet")
df.head()
```

```python
exclude_cols = ['Start_Time', 'End_Time', 'user_id', 'Depression_label', 'split']
df_features = df.drop(columns=exclude_cols, errors='ignore')
```

```python
df_features.head()
```

```python
missing = df_features.isnull().sum()
print("Missing values per column:\n", missing[missing > 0])
```

```python
zero_percent = (df_features == 0).sum() / len(df_features)
sparse_cols = zero_percent[zero_percent > 0.9]

print(f"Number of columns with more than 90% zeros: {len(sparse_cols)}")
print("Columns:")
print(sparse_cols.sort_values(ascending=False))
```

```python
df_features = df_features.drop(columns=sparse_cols.index)
```

```python
df_features.head()
```

```python
print(df_features['gender'].value_counts())
```

```python
df_encoded = pd.get_dummies(df_features, columns=['gender'], drop_first=False)
print([col for col in df_encoded.columns if 'gender' in col])
```

```python
df_encoded['gender_male'] = df_encoded['gender_male'].astype(int)
```

```python
print(df_encoded['gender_male'].head())
```

```python
print(df_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler

exclude_cols = ['gender_male','gender_female']

# Select columns to scale:
cols_to_scale = [col for col in df_encoded.columns if col not in exclude_cols]

scaler = StandardScaler()
df_encoded[cols_to_scale] = scaler.fit_transform(df_encoded[cols_to_scale])
```

```python
print(df_encoded.head())
```

```python
numeric_cols = df_encoded.select_dtypes(include=['float64', 'int64']).columns
stats = df_encoded[numeric_cols].agg(['mean', 'std']).T 
print(stats)

unscaled = stats[(stats['mean'].abs() > 0.1) | ((stats['std'] < 0.8) | (stats['std'] > 1.2))]
print("\nColumns that might NOT be scaled properly:")
print(unscaled)
```

```python
import pandas as pd

exclude_cols = ['Start_Time', 'End_Time', 'user_id', 'Depression_label', 'split']

df_original = pd.read_parquet("../data/processed/all_user_combined_data.parquet")
cols_to_add_back = df_original[exclude_cols]

df_final = pd.concat([df_encoded.reset_index(drop=True), cols_to_add_back.reset_index(drop=True)], axis=1)
df_final.to_parquet("../data/processed/all_user_combined_data_processed.parquet", index=False)
```

```python
import pandas as pd

df_loaded = pd.read_parquet("../data/processed/all_user_combined_data_processed.parquet")

print("Shape of dataframe:", df_loaded.shape)

exclude_cols = ['End_Time', 'user_id', 'Depression_label', 'split', 'Start_Time']
cols_to_check = [col for col in df_loaded.columns if col not in exclude_cols]

means = df_loaded[cols_to_check].mean()
stds = df_loaded[cols_to_check].std()

not_scaled_cols = [col for col in cols_to_check if abs(means[col]) > 1e-2 or abs(stds[col] - 1) > 1e-2]
print("\nColumns that might NOT be scaled properly:")
print(not_scaled_cols)

missing_cols = df_loaded[cols_to_check].columns[df_loaded[cols_to_check].isna().any()].tolist()
print("\nColumns with missing (NaN) values:")
print(missing_cols)

null_cols = df_loaded[cols_to_check].columns[df_loaded[cols_to_check].isnull().any()].tolist()
print("\nColumns with null values:")
print(null_cols)

print("\nFirst 5 rows:")
print(df_loaded.head())
```
