```python
import os

def list_all_files_and_folders(base_path='./depression-detection-model'):
    for root, dirs, files in os.walk(base_path):
        print(f'Current folder: {root}')
        if dirs:
            print(' Subfolders:')
            for d in dirs:
                print(f'  - {d}')
        if files:
            print(' Files:')
            for f in files:
                print(f'  - {f}')
        print('')  # blank line for readability

# Call function for current directory
list_all_files_and_folders()
```
