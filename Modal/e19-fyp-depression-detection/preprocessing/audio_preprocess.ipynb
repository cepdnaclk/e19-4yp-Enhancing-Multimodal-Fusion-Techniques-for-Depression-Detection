```python
import os
import pandas as pd
import glob

def load_audio_csv(filepath):
    delimiters = [',', ';', '\t']

    for delim in delimiters:
        try:
            # Try reading with header
            df = pd.read_csv(filepath, delimiter=delim)
            # If header looks invalid (e.g. all columns unnamed or first row looks like data), reload without header
            if any('Unnamed' in str(col) for col in df.columns) or (str(df.iloc[0,0]).lower() == "'unknown'" or str(df.iloc[0,0]).lower() == "unknown"):
                df = pd.read_csv(filepath, delimiter=delim, header=None)
                # Assign default column names col0, col1, ...
                df.columns = [f"col{i}" for i in range(df.shape[1])]
            # Heuristic for correct parsing
            if 'timeStamp' in df.columns or 'frameTime' in df.columns or df.shape[1] > 5:
                return df
        except Exception:
            continue
    # Fallback to default read with header
    df = pd.read_csv(filepath)
    if any('Unnamed' in str(col) for col in df.columns) or (str(df.iloc[0,0]).lower() == "'unknown'" or str(df.iloc[0,0]).lower() == "unknown"):
        df = pd.read_csv(filepath, header=None)
        df.columns = [f"col{i}" for i in range(df.shape[1])]
    return df

def preprocess_audio_for_subject(subject_path):
    audio_folder = os.path.join(subject_path, 'audio')
    if not os.path.isdir(audio_folder):
        print(f"ðŸš« No audio folder in {subject_path}")
        return

    processed_folder = os.path.join(audio_folder, 'processed')
    os.makedirs(processed_folder, exist_ok=True)

    audio_files = glob.glob(os.path.join(audio_folder, '*.csv'))

    if not audio_files:
        print(f"âš ï¸ No CSV files found in {audio_folder}")
        return

    for file_path in audio_files:
        filename = os.path.basename(file_path)
        df = load_audio_csv(file_path)
        df.columns = df.columns.str.strip()

        # Remove the 0th row (first row) for specific OpenSMILE files
        if filename.endswith('_OpenSMILE2.3.0_mfcc.csv') or filename.endswith('_OpenSMILE2.3.0_egemaps.csv'):
            df = df.drop(index=0).reset_index(drop=True)

        # Remove 'name' column if present
        if 'name' in df.columns:
            df = df.drop(columns=['name'])

        # Determine timestamp column according to filename rules
        lower_fname = filename.lower()

        if lower_fname.endswith('densenet201.csv') or lower_fname.endswith('vgg16.csv'):
            # Timestamp column is first column (index 0)
            time_col = df.columns[0]
        elif 'opensmile2.3.0_egemaps.csv' in lower_fname or 'opensmile2.3.0_mfcc.csv' in lower_fname:
            # Prefer columns named timeStamp or frameTime, else fallback to col 1
            if 'timeStamp' in df.columns:
                time_col = 'timeStamp'
            elif 'frameTime' in df.columns:
                time_col = 'frameTime'
            elif len(df.columns) > 1:
                time_col = df.columns[1]
            else:
                time_col = df.columns[0]
        else:
            # Default fallback: timestamp is 2nd column (index 1)
            if len(df.columns) > 1:
                time_col = df.columns[1]
            else:
                time_col = df.columns[0]

        # Rename timestamp column to 'time'
        df = df.rename(columns={time_col: 'time'})

        out_path = os.path.join(processed_folder, filename)
        df.to_csv(out_path, index=False)
        print(f"âœ… Processed {filename} saved to {out_path}")

def preprocess_all_subjects_audio(root_data_dir):
    for participant_folder in os.listdir(root_data_dir):
        if participant_folder.startswith('.'):
            continue  # skip hidden/system files

        participant_path = os.path.join(root_data_dir, participant_folder)
        if os.path.isdir(participant_path):
            print(f"\nðŸ“‚ Processing audio for participant: {participant_folder}")
            preprocess_audio_for_subject(participant_path)
        else:
            print(f"ðŸš« {participant_folder} is not a folder, skipping.")

if __name__ == "__main__":
    root_data_dir = "../data"  # Change this to your root data folder path
    preprocess_all_subjects_audio(root_data_dir)
```

```python
import os
import pandas as pd

# Base path relative to your script/notebook
data_dir = os.path.join('..', 'data')

# List all subject folders that end with _P
subject_folders = [
    folder for folder in os.listdir(data_dir)
    if os.path.isdir(os.path.join(data_dir, folder)) and folder.endswith('_P')
]

def drop_first_col_if_col0(df):
    if df.columns[0] == 'col0':
        df = df.drop(df.columns[0], axis=1)
    return df

for subject in sorted(subject_folders):
    audio_dir = os.path.join(data_dir, subject, 'audio', 'processed')
    print(f"\n=== Audio CSV files for subject {subject} ===")

    if not os.path.exists(audio_dir):
        print(f"âŒ Audio folder not found for {subject} at path: {audio_dir}")
        continue

    # List all CSV files in audio folder
    audio_csv_files = [f for f in os.listdir(audio_dir) if f.endswith('.csv')]

    if not audio_csv_files:
        print(f"âš ï¸ No CSV files found in audio folder of {subject}")
        continue

    for csv_file in sorted(audio_csv_files):
        csv_path = os.path.join(audio_dir, csv_file)
        print(f"\nðŸ”¹ Processing {csv_file} ...")
        try:
            df = pd.read_csv(csv_path)
            # Drop first column if named 'col0'
            df = drop_first_col_if_col0(df)
            # Save back to same path, overwriting original
            df.to_csv(csv_path, index=False)
            print(f"âœ… Processed and saved {csv_file}")
        except Exception as e:
            print(f"âŒ Failed to process {csv_file}: {e}")
```

```python
"""
This script scans through audio feature CSV files located in each subject's
'processed' audio folder under the '../data' directory. It checks whether the
numeric data columns (excluding the 'time' column) are normalized between 0 and 1.

- If the data is already normalized, it prints a confirmation.
- If not normalized, it normalizes all numeric columns (except 'time') using min-max scaling
  to scale values between 0 and 1 and saves the normalized data back to the same CSV file.

This ensures all audio feature CSV files have normalized numeric data ready for further processing.
"""

import os
import pandas as pd
import numpy as np

data_dir = os.path.join('..', 'data')

subject_folders = [
    folder for folder in os.listdir(data_dir)
    if os.path.isdir(os.path.join(data_dir, folder)) and folder.endswith('_P')
]

def is_normalized(df, exclude_cols=['time']):
    cols_to_check = [col for col in df.columns if col not in exclude_cols]
    numeric_cols = df[cols_to_check].select_dtypes(include=[np.number]).columns.tolist()
    if not numeric_cols:
        return True  # No numeric columns to check
    col_min = df[numeric_cols].min().min()
    col_max = df[numeric_cols].max().max()
    return (col_min >= 0) and (col_max <= 1)

def normalize_df(df, exclude_cols=['time']):
    cols_to_normalize = [col for col in df.columns if col not in exclude_cols]
    numeric_cols = df[cols_to_normalize].select_dtypes(include=[np.number]).columns.tolist()

    for col in numeric_cols:
        col_min = df[col].min()
        col_max = df[col].max()
        if col_max > col_min:  # avoid division by zero
            df[col] = (df[col] - col_min) / (col_max - col_min)
        else:
            df[col] = 0  # if all values same, set to zero
    return df

for subject in sorted(subject_folders):
    audio_dir = os.path.join(data_dir, subject, 'audio', 'processed')
    print(f"\n=== Checking normalization for subject {subject} ===")

    if not os.path.exists(audio_dir):
        print(f"âŒ Audio folder not found for {subject} at path: {audio_dir}")
        continue

    audio_csv_files = [f for f in os.listdir(audio_dir) if f.endswith('.csv')]

    if not audio_csv_files:
        print(f"âš ï¸ No CSV files found in audio folder of {subject}")
        continue

    for csv_file in sorted(audio_csv_files):
        csv_path = os.path.join(audio_dir, csv_file)
        try:
            df = pd.read_csv(csv_path)

            if is_normalized(df, exclude_cols=['time']):
                print(f"{csv_file}: âœ… Normalized")
            else:
                print(f"{csv_file}: âŒ NOT Normalized â€” Normalizing now...")
                df = normalize_df(df, exclude_cols=['time'])
                df.to_csv(csv_path, index=False)
                print(f"{csv_file}: âœ… Normalized and saved.")

        except Exception as e:
            print(f"âŒ Failed to process {csv_file}: {e}")
```
