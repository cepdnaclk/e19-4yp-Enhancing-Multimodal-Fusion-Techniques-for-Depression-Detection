```python
import os
import pandas as pd
import glob
import scipy.io

def load_video_csv(filepath):
    delimiters = [',', ';', '\t']

    for delim in delimiters:
        try:
            df = pd.read_csv(filepath, delimiter=delim)
            if any('Unnamed' in str(col) for col in df.columns) or \
               str(df.iloc[0, 0]).strip().lower() in ["'unknown'", "unknown"]:
                df = pd.read_csv(filepath, delimiter=delim, header=None)
                df.columns = [f"col{i}" for i in range(df.shape[1])]

            if any(c in df.columns for c in ['frame', 'timestamp', 'time', 'frameTime']):
                return df
        except Exception:
            continue
    df = pd.read_csv(filepath)
    if any('Unnamed' in str(col) for col in df.columns) or \
       str(df.iloc[0, 0]).strip().lower() in ["'unknown'", "unknown"]:
        df = pd.read_csv(filepath, header=None)
        df.columns = [f"col{i}" for i in range(df.shape[1])]
    return df

def load_video_mat(filepath):
    try:
        mat_data = scipy.io.loadmat(filepath)
        mat_keys = [k for k in mat_data.keys() if not k.startswith('__')]
        if len(mat_keys) == 0:
            print(f"⚠️ No valid data found in {filepath}")
            return None

        data_key = mat_keys[0]
        mat_array = mat_data[data_key]

        df = pd.DataFrame(mat_array)

        # Rename first two columns to frame and time if possible
        if df.shape[1] >= 2:
            df.rename(columns={0: 'frame', 1: 'time'}, inplace=True)

        return df
    except Exception as e:
        print(f"❌ Error loading {filepath}: {e}")
        return None

def preprocess_video_for_subject(subject_path):
    video_folder = os.path.join(subject_path, 'video')
    if not os.path.isdir(video_folder):
        print(f"🚫 No video folder in {subject_path}")
        return

    processed_folder = os.path.join(video_folder, 'processed')
    os.makedirs(processed_folder, exist_ok=True)

    # Process CSV files
    video_csv_files = glob.glob(os.path.join(video_folder, '*.csv'))
    for file_path in video_csv_files:
        filename = os.path.basename(file_path)
        df = load_video_csv(file_path)
        df.columns = df.columns.str.strip()

        if 'name' in df.columns:
            df = df.drop(columns=['name'])

        if df.shape[1] >= 2:
            df.rename(columns={df.columns[0]: 'frame', df.columns[1]: 'time'}, inplace=True)

        out_path = os.path.join(processed_folder, filename)
        df.to_csv(out_path, index=False)
        print(f"✅ Processed CSV {filename} saved to {out_path}")

    # Process MAT files
    video_mat_files = glob.glob(os.path.join(video_folder, '*.mat'))
    for file_path in video_mat_files:
        filename = os.path.basename(file_path)
        df = load_video_mat(file_path)
        if df is not None:
            out_filename = filename.replace('.mat', '.csv')
            out_path = os.path.join(processed_folder, out_filename)
            df.to_csv(out_path, index=False)
            print(f"✅ Processed MAT {filename} saved as CSV to {out_path}")

def preprocess_all_subjects_video(root_data_dir):
    for participant_folder in os.listdir(root_data_dir):
        if participant_folder.startswith('.'):
            continue
        participant_path = os.path.join(root_data_dir, participant_folder)
        if os.path.isdir(participant_path):
            print(f"\n📂 Processing video for participant: {participant_folder}")
            preprocess_video_for_subject(participant_path)
        else:
            print(f"🚫 {participant_folder} is not a folder, skipping.")

if __name__ == "__main__":
    root_data_dir = "../data"  # adjust path as needed
    preprocess_all_subjects_video(root_data_dir)
```

```python
import os
import glob
import pandas as pd

def remove_frame_column_from_csv_folder(folder_path):
    csv_files = glob.glob(os.path.join(folder_path, '*.csv'))
    if not csv_files:
        print(f"⚠️ No CSV files found in {folder_path}")
        return

    for file_path in csv_files:
        df = pd.read_csv(file_path)
        if 'frame' in df.columns:
            df = df.drop(columns=['frame'])
            df.to_csv(file_path, index=False)
            print(f"✅ Removed 'frame' column from {os.path.basename(file_path)}")
        else:
            print(f"ℹ️ No 'frame' column in {os.path.basename(file_path)}")

def remove_frame_from_all_participants(root_data_dir):
    for participant_folder in os.listdir(root_data_dir):
        if participant_folder.startswith('.'):
            continue  # skip hidden/system files
        participant_path = os.path.join(root_data_dir, participant_folder)
        if os.path.isdir(participant_path):
            video_processed_folder = os.path.join(participant_path, 'video', 'processed')
            if os.path.isdir(video_processed_folder):
                print(f"\n📂 Processing participant: {participant_folder}")
                remove_frame_column_from_csv_folder(video_processed_folder)
            else:
                print(f"⚠️ No 'video/processed' folder for {participant_folder}, skipping.")
        else:
            print(f"🚫 {participant_folder} is not a folder, skipping.")

if __name__ == "__main__":
    root_data_dir = "../data"  # Change this to your root data folder path
    remove_frame_from_all_participants(root_data_dir)

```

```python
"""
This script scans through video feature CSV files located in each subject's
'processed' video folder under the '../data' directory. It checks whether the
numeric data columns (excluding 'time' and 'frame' columns) are normalized between 0 and 1.

- If the data is already normalized, it prints a confirmation.
- If not normalized, it normalizes all numeric columns (except 'time' and 'frame') using min-max scaling
  to scale values between 0 and 1 and saves the normalized data back to the same CSV file.

This ensures all video feature CSV files have normalized numeric data ready for further processing.
"""

import os
import pandas as pd
import numpy as np

data_dir = os.path.join('..', 'data')

subject_folders = [
    folder for folder in os.listdir(data_dir)
    if os.path.isdir(os.path.join(data_dir, folder)) and folder.endswith('_P')
]

def is_normalized(df, exclude_cols=['time', 'frame']):
    cols_to_check = [col for col in df.columns if col not in exclude_cols]
    numeric_cols = df[cols_to_check].select_dtypes(include=[np.number]).columns.tolist()
    if not numeric_cols:
        return True  # No numeric columns to check
    col_min = df[numeric_cols].min().min()
    col_max = df[numeric_cols].max().max()
    return (col_min >= 0) and (col_max <= 1)

def normalize_df(df, exclude_cols=['time', 'frame']):
    cols_to_normalize = [col for col in df.columns if col not in exclude_cols]
    numeric_cols = df[cols_to_normalize].select_dtypes(include=[np.number]).columns.tolist()

    for col in numeric_cols:
        col_min = df[col].min()
        col_max = df[col].max()
        if col_max > col_min:  # avoid division by zero
            df[col] = (df[col] - col_min) / (col_max - col_min)
        else:
            df[col] = 0  # if all values same, set to zero
    return df

for subject in sorted(subject_folders):
    video_dir = os.path.join(data_dir, subject, 'video', 'processed')
    print(f"\n=== Checking normalization for subject {subject} ===")

    if not os.path.exists(video_dir):
        print(f"❌ Video folder not found for {subject} at path: {video_dir}")
        continue

    video_csv_files = [f for f in os.listdir(video_dir) if f.endswith('.csv')]

    if not video_csv_files:
        print(f"⚠️ No CSV files found in video folder of {subject}")
        continue

    for csv_file in sorted(video_csv_files):
        csv_path = os.path.join(video_dir, csv_file)
        try:
            df = pd.read_csv(csv_path)

            if is_normalized(df, exclude_cols=['time', 'frame']):
                print(f"{csv_file}: ✅ Normalized")
            else:
                print(f"{csv_file}: ❌ NOT Normalized — Normalizing now...")
                df = normalize_df(df, exclude_cols=['time', 'frame'])
                df.to_csv(csv_path, index=False)
                print(f"{csv_file}: ✅ Normalized and saved.")

        except Exception as e:
            print(f"❌ Failed to process {csv_file}: {e}")
```

```python
import pandas as pd
import os

# Base data directory (adjust as needed)
base_path = '../data'

# Loop through each subject folder
for subject in os.listdir(base_path):
    subject_path = os.path.join(base_path, subject)
    
    if not os.path.isdir(subject_path):
        continue  # Skip non-folder entries

    video_csv_path = os.path.join(subject_path, 'video', 'processed', f'{subject.split("_")[0]}_CNN_ResNet.csv')
    audio_csv_path = os.path.join(subject_path, 'audio', 'processed', f'{subject.split("_")[0]}_densenet201.csv')

    # Check if both files exist
    if not os.path.isfile(video_csv_path) or not os.path.isfile(audio_csv_path):
        print(f"Skipping {subject}: Required files not found.")
        continue

    try:
        video_df = pd.read_csv(video_csv_path)
        audio_df = pd.read_csv(audio_csv_path)

        # Check for 'time' column
        if 'time' not in video_df.columns or 'time' not in audio_df.columns:
            print(f"Skipping {subject}: 'time' column missing.")
            continue

        # Extend audio time column
        audio_time = audio_df['time'].copy()
        if len(audio_time) < len(video_df):
            last_value = audio_time.iloc[-1]
            extra_rows = len(video_df) - len(audio_time)
            audio_time = audio_time.tolist() + [last_value] * extra_rows
        else:
            audio_time = audio_time[:len(video_df)]

        # Rename and insert time columns
        video_df = video_df.rename(columns={'time': '1'})
        video_df.insert(0, 'time', audio_time)

        # Save the updated CSV
        video_df.to_csv(video_csv_path, index=False)
        print(f"Updated: {subject}")

    except Exception as e:
        print(f"Error processing {subject}: {e}")
```
